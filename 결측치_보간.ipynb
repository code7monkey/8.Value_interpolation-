```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

# Load the data
data_path = '/content/drive/MyDrive/결측치 보간/data.csv'
data = pd.read_csv(data_path)
```

```python
# Linear interpolation
linear_interpolated_data = data.copy()
linear_interpolated_data['Value'] = linear_interpolated_data['Value'].interpolate(method='linear')

# Polynomial interpolation (order=2 for simplicity)
poly_interpolated_data = data.copy()
poly_interpolated_data['Value'] = poly_interpolated_data['Value'].interpolate(method='polynomial', order=2)

# Spline interpolation (order=3)
spline_interpolated_data = data.copy()
spline_interpolated_data['Value'] = spline_interpolated_data['Value'].interpolate(method='spline', order=3)

# KNN interpolation with increased neighbors (testing with 8)
knn_imputer = KNNImputer(n_neighbors=8)
knn_interpolated_data = data.copy()
knn_interpolated_data['Value'] = knn_imputer.fit_transform(knn_interpolated_data[['Value']])
```

```python
# Weights can be adjusted based on performance of each method
weights = [0.88, 0.04, 0.04, 0.04]  # Adjusting weights for linear, poly, spline, and KNN
total_weight = sum(weights)

ensemble_interpolated_data = data.copy()
ensemble_interpolated_data['Value'] = (
    weights[0] * linear_interpolated_data['Value'] +
    weights[1] * poly_interpolated_data['Value'] +
    weights[2] * spline_interpolated_data['Value'] +
    weights[3] * knn_interpolated_data['Value']
) / total_weight

# Save the ensemble interpolated data and download it directly
output_path = 'ensemble_interpolated_data.csv'
ensemble_interpolated_data.to_csv(output_path, index=False)

from google.colab import files
files.download(output_path)
```
